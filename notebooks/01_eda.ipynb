```python
df = df[df["loan_status"].isin(["Fully Paid", "Charged Off", "Default","Late(31-120 days)", "Late (16-30 days)"])] #removed current and issued
```

```python
df["loan_status"].value_counts()
```

```python
df["target"] = df["loan_status"].apply(lambda x: 0 if x == "Fully Paid" else 1)

df["target"].value_counts() #divided fully paid into "good loan" and others into "bad loan" and converted good loan into 0 and others into 1
```

```python
# 0 = good customers
# 1 = risky customers
```

```python
important_columns = [
    "loan_amnt",
    "term",
    "int_rate",
    "grade", 
    "sub_grade",
    "emp_length",
    "home_ownership",
    "annual_inc",
    "purpose",
    "dti",
    "fico_range_low",
    "revol_bal",
    "total_acc",
    "target"
]

df = df[important_columns]

df.head() #dti is debt to income ratio (how much debt a person has compared to their income
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
df["emp_length"] = df["emp_length"].fillna(df["emp_length"].mode()[0], inplace = True)
df["dti"] = df["dti"].fillna(df["dti"].median(), inplace = True)
```
